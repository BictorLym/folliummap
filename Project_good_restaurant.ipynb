```python
착한식당 = pd.read_csv('행자부_추천(test)(csv).csv', encoding = 'cp949')
현지인식당 = pd.read_csv('현지인_추천(test)(csv).csv', encoding = 'cp949')
```

```python
m = folium.Map(
    location = [35.08334457377809, 129.06119070455802],
    zoom_start = 13
)

tooltip = 'Click!!'

for i in range(착한식당.shape[0]):
    folium.Marker(
        [착한식당.iloc[i]['위도'], 착한식당.iloc[i]['경도']],
        popup = f'<div style = width:100px><strong>{착한식당.iloc[i]["업소명"]}</strong>',
        tooltip = tooltip,
        icon = folium.Icon(color='red')
    ).add_to(m)

for i in range(현지인식당.shape[0]):
    folium.Marker(
        [현지인식당.iloc[i]['위도'], 현지인식당.iloc[i]['경도']],
        popup = f'<div style = width:100px><strong>{현지인식당.iloc[i]["사용장소(가맹점명)"]}</strong>',
        tooltip = tooltip,
        icon = folium.Icon(color='green')
    ).add_to(m)

m
```
